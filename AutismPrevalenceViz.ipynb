```python
import pandas as pd
import plotly.express as px
import numpy as np
```

```python
data = pd.read_csv("/content/IdentifiedPrevASD.csv")
```

```python
data.head()
```

```python
fig = px.scatter(data, x = 'Surveillance Year', y = 'Combined Prevalence per 1,000 Children', color_discrete_sequence = ['green'], trendline ='ols' )

fig.update_layout(
    yaxis_title_text='Combined Prevalence per 1,000 Children',
    font_size=14,
    title_font_color = "black")

fig.show()
```
